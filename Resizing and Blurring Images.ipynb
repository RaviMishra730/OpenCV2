# Image Rotation

```python
rows, cols = img_rgb.shape[:-1]
```

```python
Matrix = cv2.getRotationMatrix2D((rows/2, cols/2),10,1) # Here 10 is the angle of rotation and 1 is the scale up and scale down 
# parameter
```

```python
image = cv2.warpAffine(img_rgb, Matrix, (cols, rows))
```

```python
plt.imshow(image)
```

```python
Matrix_2 = cv2.getRotationMatrix2D((rows/2, cols/2),10,1.3)
```

```python
image2 = cv2.warpAffine(img_rgb, Matrix_2, (cols, rows))
```

```python
plt.imshow(image2)
```

```python
 Matrix_3 = cv2.getRotationMatrix2D((rows/2, cols/2),10,0.8)
```

```python
image3 = cv2.warpAffine(img_rgb, Matrix_3, (cols, rows))
```

```python
plt.imshow(image3)
```

## in order  to rotate the image in clockwise direction

```python
 Matrix_3 = cv2.getRotationMatrix2D((rows/2, cols/2),-9,1)
```

```python
image3 = cv2.warpAffine(img_rgb, Matrix_3, (cols, rows))
```

```python
plt.imshow(image3)
```

# Image Translation

# Edge Detection

##  Algorithm -> Canny Edge Detection

```python
i = cv2.Canny(img_rgb, 10, 10) # 10, 10 is very less threshold value, if we increase the value then it will be detect the 
# edges more accurately 
```

```python
plt.imshow(i)
```

```python
i2 = cv2.Canny(img_rgb, 60, 60) # Note -> There has to be a proper threshold it's not like we can give any bigger value we want
```

```python
print(plt.imshow(i2))
```
